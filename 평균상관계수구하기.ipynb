```python
fp1=np.polyfit(average_daehak['점포'],average_daehak['평균매출액'],1) #여기서 종속변수를 돌아가면서 대입했습니다.
fp1
```

```python
f1=np.poly1d(fp1)
print(f1)
fx=np.linspace(0,500,10)
print(fx)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(average_daehak['점포'],average_daehak['평균매출액'], s=50)
plt.plot(fx,f1(fx),ls='dashed', lw=3,color='g')
plt.xlabel('점포')
plt.ylabel('평균매출액')
plt.grid()
plt.show()
```
